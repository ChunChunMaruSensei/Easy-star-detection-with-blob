```python
from skimage.feature import blob_log
from skimage.io import imread
from google.colab.patches import cv2_imshow
import cv2
from google.colab import drive
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
drive.mount('/content/drive/')

#path = '/content/drive/MyDrive/image processing'

star_image = imread('/content/drive/MyDrive/image processing/star2.jpg') #import image
star_image_gray = cv2.cvtColor(star_image, cv2.COLOR_RGB2GRAY) #convert rgb image to gray scale image
blob_data = blob_log(star_image, max_sigma=30, num_sigma=10, threshold=.1) #use blob to detect the star

fig, ax = plt.subplots(1)
ax.imshow(star_image_gray, cmap='gray', vmin=0, vmax=255)
ax.axis('off')

for i in blob_data: #loop draw circle in a picture by using the location from blob_data
  circ = Circle((i[1], i[0],),i[2], color='yellow',linewidth=2, fill=False) #blob_data => [y axis, x axis, radius] 
  ax.add_patch(circ)

img_resize = cv2.resize(star_image_gray, (304,231))
cv2_imshow(img_resize)
plt.show()
count_star = len(blob_data)
print("number of stars : %d"%(count_star))



```
